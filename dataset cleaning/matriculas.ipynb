# Select Relevant Variables

```python
relevant_columns = [
    "cod_centro",
    "des_centro",
    "cod_titulacion",
    "des_titulacion",
    "anio_nacimiento",
    "des_genero",
    "des_pais_nacionalidad",
    "des_continente_nacionalidad",
    "des_dedicacion",
    "num_total_creditos_mat_curso",
    "num_total_creditos_sup_curso",
    "num_total_creditos_no_sup_curso",
    "ind_se_titula_curso",
]

matriculas_clean = matriculas[relevant_columns].copy()

matriculas_clean.shape
```

```python
matriculas_clean["cod_centro"].dtype
```

One thing to note is that we decided to check the data type of `cod_centro` because it is an identifier that must match exactly with other tables. We found out that it was detected as float64, which produced   precision errors when we downloaded the dataset into DBeaver since floats store approximate values. Because of this, some center codes were rounded incorrectly and it broke the relationships between datasets. Therefore, we came back to modify the code to change it to an integer type to ensure consistency. This will be modified after the missing values are removed in the following section.

# Remove Rows Without Enrolled Credits

```python
matriculas_clean = matriculas_clean[matriculas_clean["num_total_creditos_mat_curso"].notnull()]
matriculas_clean = matriculas_clean.reset_index(drop=True)

print("New shape:", matriculas_clean.shape)
```

# Remove Zero Enrolled Credits

```python
matriculas_clean = matriculas_clean[matriculas_clean["num_total_creditos_mat_curso"] > 0]
matriculas_clean = matriculas_clean.reset_index(drop=True)

print("Shape after removing zero credits:", matriculas_clean.shape)
```

```python
matriculas_clean["progression_rate"] = (
    matriculas_clean["num_total_creditos_sup_curso"] /
    matriculas_clean["num_total_creditos_mat_curso"]
)

matriculas_clean["progression_rate"].describe()
```

# Remove Rows With No Center

In the section "Check for Missing Values", some records presented missing values in the column `cod_centro`, so to see if we could recover these values, we examined the rows and attempted to match them with their corresponding degrees. However, we found out that these rows contained missing values across multiple key attributes. Therefore, we decided to remove these incomplete records from the dataset to ensure data quality and consistency.

```python
matriculas_clean = matriculas_clean.dropna(subset=["cod_centro"])
```

```python
matriculas_clean["cod_centro"].isnull().sum()
```

```python
matriculas_clean["cod_centro"] = matriculas_clean["cod_centro"].astype(int)
```
