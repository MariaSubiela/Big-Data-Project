```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_excel("/content/uc3m-departamentos-2023-24.xlsx")
df.head()
print(df.shape)
print(df.columns.tolist())
```

# Remove unnecessary columns

```python
cols_to_drop = [
    "cod_universidad",
    "cod_municipio",
    "cod_situacion_departamento",
    "fecha_desde_situacion_departamento",




]

df = df.drop(columns=cols_to_drop, errors="ignore")
print(df.shape)
df.head()
```

# Add index

```python
campus_map = {
    "Leganés": 1,
    "Getafe": 2,
    "Colmenarejo": 3,
    "Madrid- Puerta de Toledo": 4,
    "Aranjuez": 5,
    "Intercampus": 6
}

df["cod_campus"] = df["des_campus"].map(campus_map)

df[["des_campus","cod_campus"]].head()
```

# null values check

```python

null_counts = df.isnull().sum()

print("Null values per column:")
print(null_counts)
```

# Department distribution

```python

dept_counts = df["des_departamento"].value_counts()

dept_table = pd.DataFrame({
    "Department": dept_counts.index,
    "Number of Records": dept_counts.values,

})

dept_table
```

# Campus distribution

```python

campus_counts = df["des_campus"].value_counts()

campus_table = pd.DataFrame({
    "Campus": campus_counts.index,
    "Number of Departments": campus_counts.values,

})

campus_table
```

# Department per campus distribution

```python
pd.crosstab(df["des_campus"], df["des_departamento"])
```

# Save clean dataset

```python

output_path = "departamentos_limpio.xlsx"

df.to_excel(output_path, index=False)

print("File saved as:", output_path)
from google.colab import files
files.download("departamentos_limpio.xlsx")
```
