# Inspect structure

```python
print("Shape:", departamentos.shape)

print("\nColumns:")
print(departamentos.columns)

print("\nMissing values:")
print(departamentos.isnull().sum())
```

```python
departamentos.info()
```

# Standardizing campus identifiers

```python
campus_map = {
    3601: 1,  # Leganés
    3602: 2,  # Getafe
    3603: 3,  # Colmenarejo
    3604: 4,  # Madrid
    3605: 5   # Aranjuez
}

departamentos["campus_id"] = departamentos["cod_campus"].map(campus_map)

departamentos["campus_id"] = departamentos["campus_id"].fillna(6)

departamentos[["cod_campus", "des_campus", "campus_id"]].head()
```

# Relevant columns

```python
cols_keep = [
    "cod_departamento",
    "des_departamento",
    "campus_id",
    "des_situacion_departamento",
]

departamentos_clean = departamentos[cols_keep].copy()

departamentos_clean.head()
```
