# Upload data

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
acceso = pd.read_csv("/content/uc3m-acceso-2023-24-anonimizado.xlsx - Sheet1.csv")

acceso.head()
```

# Inspect structure

```python
print("Shape:", acceso.shape)

print("\nColumns:")
print(acceso.columns)

print("\nMissing values:")
print(acceso.isnull().sum())
```

```python
acceso.info()
```

# Relevant columns

```python
cols_keep = [
    "curso_academico",
    "cod_centro",
    "des_centro",
    "cod_titulacion",
    "des_titulacion",
    "nota_admision",
    "des_forma_admision",
    "anio_nacimiento",
    "des_genero",
    "des_nivel_estudios_madre",
    "des_ocupacion_madre",
    "des_nivel_estudios_padre",
    "des_ocupacion_padre"
]

acceso_clean = acceso[cols_keep].copy()

acceso_clean.head()
```

# Remove rows with key missing values

```python
acceso_clean = acceso_clean.dropna(subset=["nota_admision"])
```

### Missing values in parent information were filled with "unknown"

```python
acceso_clean["des_ocupacion_madre"] = acceso_clean["des_ocupacion_madre"].fillna("Unknown")
acceso_clean["des_ocupacion_padre"] = acceso_clean["des_ocupacion_padre"].fillna("Unknown")
acceso_clean["des_nivel_estudios_madre"] = acceso_clean["des_nivel_estudios_madre"].fillna("Unknown")
acceso_clean["des_nivel_estudios_padre"] = acceso_clean["des_nivel_estudios_padre"].fillna("Unknown")
```

# Convert cod_centro to integer

```python
acceso_clean["cod_centro"] = acceso_clean["cod_centro"].astype("Int64")
```

# Basic statistics

### Admission score

```python
acceso_clean["nota_admision"].describe()
```

### Student age

```python
acceso_clean["anio_nacimiento"].describe()
```

# Categorical analysis

### Gender distribution

```python
acceso_clean["des_genero"].value_counts()
```

### Center distribution

```python
acceso_clean["des_centro"].value_counts()
```

### Degree distribution

```python
acceso_clean["des_titulacion"].value_counts()
```

# Visualisations

### Gender distribution

```python
acceso_clean["des_genero"].value_counts().plot(kind="bar")

plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

### Center distribution

```python
acceso_clean["des_centro"].value_counts().plot(kind="bar")

plt.title("Students per Campus")
plt.xlabel("Campus")
plt.ylabel("Count")
plt.show()
```

### Admission score distribution

```python
plt.hist(acceso_clean["nota_admision"], bins=20)

plt.title("Admission Score Distribution")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```

# Average admission score by degree

```python
acceso_clean.groupby("des_titulacion")["nota_admision"].mean().sort_values(ascending=False)
```

# Average admission score by center

```python
acceso_clean.groupby("des_centro")["nota_admision"].mean().sort_values(ascending=False)
```

# Average score by gender

```python
acceso_clean.groupby("des_genero")["nota_admision"].mean()
```

# Create an ID column

```python
acceso_clean = acceso_clean.reset_index(drop=True)

acceso_clean.insert(0, "access_id", acceso_clean.index + 1)
```

# Export cleaned dataset

```python
acceso_clean.to_csv("acceso.csv", index=False)
```
