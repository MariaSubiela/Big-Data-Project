```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("moventrante_limpio.xlsx")

print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())

df.head()
```

Add index

```python
mapping = {
    "LEGANÉS": 1,
    "GETAFE": 2,
    "COLMENAREJO": 3,
    "MADRID": 4,
    "ARANJUEZ": 5,
    "INTERCAMPUS": 6
}

df["campus_id"] = df["des_municipio_destino"].str.upper().map(mapping)
df.head()
```

# Check null values

```python
# Count of null values per column
null_counts = df.isnull().sum()

print("Number of null values per column:")
print(null_counts)
```

# Mobility type per student

```python
program_counts = df["des_tipo_programa_mov"].value_counts()

print(program_counts)

plt.figure()
program_counts.plot(kind="bar")
plt.title("Incoming Students by Mobility Program Type")
plt.ylabel("Number of Students")
plt.show()
```

# Mobility program per student

```python
mobility_program = df["des_programa_movilidad"].value_counts()

print(mobility_program)

plt.figure()
mobility_program.plot(kind="bar")
plt.title("Incoming Students by Mobility Program")
plt.ylabel("Number of Students")
plt.show()
```

# Gender distribution

```python
gender_counts = df["des_genero"].value_counts()

print(gender_counts)

plt.figure()
gender_counts.plot(kind="pie", autopct="%1.1f%%")
plt.title("Gender Distribution of Incoming Students")
plt.ylabel("")
plt.show()
```

# Most common nationalities

```python
nationality_counts = df["des_pais_nacionalidad"].value_counts().head(15)

print(nationality_counts)

plt.figure()
nationality_counts.plot(kind="bar")
plt.title("Top Nationalities of Incoming Students")
plt.ylabel("Number of Students")
plt.show()
```

# Campus destination

```python
municipality_counts = df["des_municipio_destino"].value_counts()

print(municipality_counts)

plt.figure()
municipality_counts.plot(kind="bar")
plt.title("Incoming Students by Destination Municipality")
plt.ylabel("Number of Students")
plt.show()
```
