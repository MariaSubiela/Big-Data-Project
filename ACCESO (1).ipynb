# Create the mapping so that we have the same index across all datasets

```python
campus_map = {
    "Leganés": 1,
    "Getafe": 2,
    "Colmenarejo": 3,
    "Madrid": 4,
    "Aranjuez": 5,
    "Intercampus": 6
}
```

```python
df["cod_centro"] = df["des_campus"].map(campus_map)
```

```python
df[["des_campus", "cod_centro"]].drop_duplicates()
```

# Check for null values

```python
null_counts = df.isnull().sum()
null_counts[null_counts > 0]
```

# Students per campus

```python
df["des_campus"].value_counts()
df["des_campus"].value_counts().plot(kind="bar")
plt.title("Distribution of Students by Campus")
plt.xlabel("Campus")
plt.ylabel("Number of Students")
plt.show()
```

# Gender distribution

```python
df["des_genero"].value_counts()
df["des_genero"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()
```
