```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_excel("/content/uc3m-centros-2023-24.xlsx")
df.head()
print(df.shape)
print(df.columns.tolist())
```

# Drop unnecessary columns

```python
cols_to_drop = [
    "cod_universidad",
    "cod_centro",
    "cod_municipio",
    "cod_campus",
    "cod_tipo_centro",
    "cod_naturaleza_centro",
    "fecha_desde_situacion_centro",
    "latitude",
    "longitude",
    "cod_situacion_centro"



]

df = df.drop(columns=cols_to_drop, errors="ignore")
print(df.shape)
df.head()
```

# Add campus index

```python
campus_map = {
    "Leganés": 1,
    "Getafe": 2,
    "Colmenarejo": 3,
    "Madrid- Puerta de Toledo": 4,
    "Aranjuez": 5,
    "Intercampus": 6
}

df["id_campus"] = df["des_campus"].map(campus_map)

df[["des_campus","id_campus"]].head()
```

```python
df.head()
```

# Null values check

```python
# Count null values per column
null_counts = df.isnull().sum()

print("Null values per column:")
print(null_counts)
```

# Campus distribution

```python
df["des_campus"].value_counts()
```

# Type of center distribution

```python
df["des_tipo_centro"].value_counts()
df["des_tipo_centro"].value_counts().plot(kind="bar")

plt.title("Distribution of Center Types")
plt.xlabel("Type of Center")
plt.ylabel("Number of Centers")

plt.show()
```

# Save new excel file

```python
# Save cleaned dataset to new Excel file
output_path = "centros_limpio.xlsx"

df.to_excel(output_path, index=False)

print("File saved as:", output_path)
from google.colab import files
files.download("centros_limpio.xlsx")
```
