# Select Relevant Variables

```python
relevant_columns = [
    "curso_academico",
    "cod_campus", "des_campus",
    "cod_titulacion", "des_titulacion",
    "cod_centro", "des_centro",
    "anio_nacimiento",
    "des_genero",
    "des_pais_nacionalidad",
    "des_continente_nacionalidad",
    "cod_dedicacion", "des_dedicacion",
    "num_total_creditos_mat_curso",
    "num_total_creditos_sup_curso",
    "num_total_creditos_no_sup_curso",
    "ind_se_titula_curso",
]

df_clean = df[relevant_columns].copy()

df_clean.shape
```

# Analyze Missing Values in Relevant Dataset

```python
df_clean.isnull().sum().sort_values(ascending=False)
```

```python
df_clean.describe()
```

# Remove Rows Without Enrolled Credits

```python
df_clean = df_clean[df_clean["num_total_creditos_mat_curso"].notnull()]
df_clean = df_clean.reset_index(drop=True)

print("New shape:", df_clean.shape)
```

# Remove Zero Enrolled Credits

```python
df_clean = df_clean[df_clean["num_total_creditos_mat_curso"] > 0]
df_clean = df_clean.reset_index(drop=True)

print("Shape after removing zero credits:", df_clean.shape)
```

```python
df_clean["progression_rate"] = (
    df_clean["num_total_creditos_sup_curso"] /
    df_clean["num_total_creditos_mat_curso"]
)

df_clean["progression_rate"].describe()
```

# Check for Impossible Values (Safety Check)

```python
df_clean[df_clean["progression_rate"] > 1].shape
df_clean[df_clean["progression_rate"] < 0].shape
```

# Investigate the Mass at 1

```python
(df_clean["progression_rate"] == 1).mean()
```

72.97% of students pass 100% of their credits.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df_clean["progression_rate"], bins=20)
plt.xlabel("Progression Rate")
plt.ylabel("Number of Students")
plt.title("Distribution of Academic Progression Rate")
plt.show()
```

1) Zoom Into Non-Perfect Students

```python
plt.figure(figsize=(8,5))
plt.hist(df_clean[df_clean["progression_rate"] < 1]["progression_rate"], bins=20)
plt.xlabel("Progression Rate (Less than 1)")
plt.ylabel("Number of Students")
plt.title("Distribution of Non-Perfect Progression")
plt.show()
```

2) Density Plot

```python
df_clean["progression_rate"].plot(kind="kde")
plt.title("Density of Academic Progression Rate")
plt.show()
```

# Check Relationship Between Enrolled and Passed Credits

```python
df_clean[["num_total_creditos_mat_curso",
          "num_total_creditos_sup_curso"]].describe()
```

# Add Index

```python
df_clean = df_clean.reset_index(drop=True)
df_clean.insert(0, "student_id", df_clean.index + 1)
```

# Campus Match for the cod_campus column

```python
df_clean["des_campus"].unique()
```

```python
df_clean = df_clean.dropna(subset=["des_campus"])
```

```python
# mapping campus names to IDs
campus_map = {
    "Leganés": 1,
    "Getafe": 2,
    "Colmenarejo": 3,
    "Madrid- Puerta de Toledo": 4,
    "Aranjuez": 5,
    "Intercampus": 6
}

# create the column based on des_campus
df_clean["cod_campus"] = df_clean["des_campus"].map(campus_map)

# ensure the values are integers, not decimals
df_clean["cod_campus"] = df_clean["cod_campus"].astype(int)
```

# Save New Dataset

```python
df_clean.to_csv("matriculas_clean.csv", index=False)
```
